### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries

import random 

#Funtion to roll the dice.
iteraciones = 1000
number_of_time = 0

for i in range (iteraciones):
    n = random.randint(1,6) + random.randint(1,6)
    if (n>7) or (n%2==0):
        number_of_time +=1
probab = round((number_of_time/iteraciones)*100, 2)
print ('La probabilidad que el numero sea mayor que 7 o par:',str(probab) +'%' )
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import random 
import numpy as np 

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  
  
        
#Initialize important variables

n_simulations = 100000
cantidad_5_iguales = 0
cantidad_3w_2r = 0



for i in range(n_simulations):
    
    #make a list of the colors that we choose
   
    lista_de_colores = []
    for j in range(5):
        lista_de_colores.append(d[random.randint(0,59)])
        
    #Convert it to a numpy array
    array_lista = np.array(lista_de_colores)

    #find the number of each taht we picked
    cant_white = sum(array_lista == 'White')
    cant_red = sum(array_lista == 'Red')
    cant_green = sum(array_lista == 'Green')
 
    #Keep track if the combination met the criteria
    
    if (cant_white == 5) or (cant_red == 5) or (cant_green == 5):
        cantidad_5_iguales += 1
    elif (cant_white == 3) and (cant_red == 2):
        cantidad_3w_2r += 1
    
prob_3w_2r = round((cantidad_3w_2r/n_simulations)*100 , 2)
prob_5_iguales = round((cantidad_5_iguales/n_simulations)*100 , 2) 

#print 'the probability of 3 white and 2 red is' in percentage
print ('La probabildiad de que 3 blancas y 2 rojas es:', str(prob_3w_2r), '%')

#print 'the probability of all the same color is' in percentage
print ('La probabilidad de que 5 sean igulaes:',str(prob_5_iguales), '%')   
```
